```python
from sklearn.datasets import make_regression
import numpy as np
import scipy as sp
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn as sk
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib as mpl
```

```python
#read all
wifi_raw = pd.read_csv("TD.csv")
wifi1 = pd.read_csv("wifi_nan.txt")
wifi00 = pd.read_csv("wifi00.txt")
wifi01 = pd.read_csv("wifi01.txt")
wifi02 = pd.read_csv("wifi02.txt")
wifi03 = pd.read_csv("wifi03.txt")
wifi04 = pd.read_csv("wifi04.txt")
wifi=wifi1
```

```python
for i in wifi_nan.iloc[:,0:520]:
    wifi_nan[i][wifi_nan[i] == 100]=np.nan
```

```python
# 각 층을 다른 파일에 저장해 주는 명령어

wifi04 = pd.DataFrame()
for i in wifi1:
    wifi04[i] = wifi1[i][wifi1["FLOOR"] == 4]

wifi00.to_csv("wifi00.txt")
wifi01.to_csv("wifi01.txt")
wifi02.to_csv("wifi02.txt")
wifi03.to_csv("wifi03.txt")
wifi04.to_csv("wifi04.txt")
```

```python
#draw total floor data of column(wifi machine number)
norm = mpl.colors.Normalize(vmin = 6, vmax=110 )
def draw_wh(column_name):
    plt.figure(figsize=(8,6))
    plt.scatter(wifi_raw["LONGITUDE"], wifi_raw["LATITUDE"], s=400,c="green", alpha=1, marker="D")
    plt.scatter(wifi["LONGITUDE"], wifi["LATITUDE"], s=50, c=(wifi[column_name]+110),vmin=wifi[column_name].min()+110,vmax=wifi[column_name].max()+110, alpha=0.7, cmap="hot")
    plt.xlim(wifi["LONGITUDE"].min(),wifi["LONGITUDE"].max())
    plt.ylim(wifi["LATITUDE"].min(),wifi["LATITUDE"].max())
    plt.show()
#draw each floor data of column(wifi machine number)
def draw_ef(column_name):
    plt.figure(figsize=(8,30))
    x = [wifi00, wifi01, wifi02, wifi03, wifi04]
    for i in range(len(x)):
        plt.subplot(5,1,i+1)
        plt.scatter(wifi_raw["LONGITUDE"], wifi_raw["LATITUDE"], s=400,c="green", alpha=1,marker="D")
        plt.scatter(x[i]["LONGITUDE"], x[i]["LATITUDE"], s=50, c=(x[i][column_name]+110),vmin=wifi[column_name].min()+110,vmax=wifi[column_name].max()+110, alpha=0.7, cmap="hot")
        plt.xlim(wifi["LONGITUDE"].min(),wifi["LONGITUDE"].max())
        plt.ylim(wifi["LATITUDE"].min(),wifi["LATITUDE"].max())
    plt.show()
#draw qq-plot vs uniform for column(wifi machine number)
def drawsignal(columns):
    sp.stats.probplot(wifi1[columns]+110, dist="uniform", plot=plt)
    plt.show()
#draw all of upper case
def draw_all(columns):
    drawsignal(columns)
   
    draw_wh(columns)
    draw_ef(columns)
```
